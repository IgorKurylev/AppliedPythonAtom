```python
PATH_TO_BD = "./mydatabase_pre.db" 
conn = sqlite3.connect(PATH_TO_BD) # или :memory: чтобы сохранить в RAM
cursor = conn.cursor()

# Создание таблицы
cursor.execute("""CREATE TABLE store
                  (Производитель text, Тип text, Code int, Скорость int, 
                   Память int, Цена int)
               """)
# Сохраняем изменения
conn.commit()

prods = [('Америка', 'Компьютер', '1', '1100', '1000', '25000'),
          ('Африка', 'Компьютер', '2', '1500', '2000', '25000'),
          ('Россия', 'Компьютер', '3', '1500', '2500', '28000'),
          ('Америка', 'Компьютер', '6', '2000', '2500', '29000'),
          ('Америка', 'Компьютер', '9', '2000', '3000', '32000')]

# выполняем сразу много вставок используя прелести "шаблончиков" sqlite3  
cursor.executemany("INSERT INTO store VALUES (?,?,?,?,?,?)", prods)

# Сохраняем изменения
conn.commit()
    
# Закрываем соединение
conn.close()
```
